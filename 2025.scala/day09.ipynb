```python
import matplotlib.pyplot as plt
```

```python
with open("input09.txt", "r") as f:
    data = [[int(i) for i in line.split(",")] for line in f.readlines()]
```

```python
def plot_rectangle(index1: int, index2: int, data: list[list[int]]):
    (x1, y1), (x2, y2) = data[index1], data[index2]
    plt.plot([x1, x1, x2, x2, x1], [y1, y2, y2, y1, y1], color="red")
```

```python
plt.plot([point[0] for point in data], [point[1] for point in data])
plot_rectangle(218, 248, data)
```
